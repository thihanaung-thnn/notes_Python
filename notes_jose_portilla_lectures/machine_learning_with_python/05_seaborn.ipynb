# Seaborn

## Scatter plots

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('dm_office_sales.csv')
```

```python
df.info()
```

```python
sns.scatterplot(x='salary', y='sales', data=df)
```

```python
# connect with matplotlib
plt.figure(figsize=(10,6))
sns.scatterplot(x='salary',y='sales',data=df)
```

### parameters
- hue
- size
- style

```python
# hue
plt.figure(figsize=(8,5))
sns.scatterplot(x='salary', y='sales', data=df, hue='division')
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='salary',y='sales',data=df,hue='work experience')
```

Choosing a palette from Matplotlib's cmap: https://matplotlib.org/tutorials/colors/colormaps.html

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='salary',y='sales',data=df,hue='work experience', palette='viridis')
```

```python
# size
plt.figure(figsize=(8,6))
sns.scatterplot(x='salary',y='sales',data=df,size='work experience')
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='salary',y='sales',data=df,s=200)
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='salary',y='sales',data=df,style='level of education')
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='salary',y='sales',data=df,style='level of education',hue='level of education', s=200)
# plt.savefig('example.jpg')
```

### Pairgrid

```python
g = sns.PairGrid(df_sp)
g = g.map_upper(sns.scatterplot) # at position (0,1),(0,2),(1,2) --zero index
g = g.map_diag(sns.kdeplot, lw=2)
g = g.map_lower(sns.kdeplot, colors='red')
```

```python
g = sns.PairGrid(df_sp, hue='gender', palette='viridis', hue_kws={'marker':['o','+']})
g = g.map_upper(sns.scatterplot, linewidths=1, edgecolor='w', s=40)
# g = g.map_diag(sns.displot)
g = g.map_lower(sns.kdeplot)
g = g.add_legend();
```

### FacetGrid

```python
sns.FacetGrid(data=df_sp, col='gender', row='lunch')
# this creates grids
```

```python
g = sns.FacetGrid(data=df_sp, col='gender', row='lunch')
g = g.map(plt.scatter, 'math score', 'reading score', edgecolor='w')
g.add_legend()
```

```python
g = sns.FacetGrid(data=df_sp, col='gender', row='lunch')
g = g.map(plt.scatter, 'math score', 'reading score', edgecolor='w')
g.add_legend()
plt.subplots_adjust(hspace=0.4, wspace=1)
```
