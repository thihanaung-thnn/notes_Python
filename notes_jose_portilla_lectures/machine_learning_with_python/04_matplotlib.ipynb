```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
```

## Basics

## Figure Object

```python
x = np.linspace(0,11,10)
y = x**4
a = np.arange(0,10)
b = a*2
```

```python
# creating a figure
# create blank canvas 
fig = plt.figure()

# Add set of axes to figure
axes = fig.add_axes([0,0,1,1]) # left, bottom, width, height

print(type(fig))
print(type(axes))

# plot on that axes
axes.plot(x,y)
plt.show()
```

```python
# add another set of axes to figure
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2, 0.5, 0.25, 0.25]) # smaller figure

ax1.plot(x,y)
ax1.set_xlabel('X label')
ax1.set_ylabel('Y label')
ax1.set_title('Big Figure')

# insert ax2 figrue
ax2.plot(x,y)
ax2.set_xlim(8,10)
ax2.set_ylim(4000,10000)
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('Zoomed In')
```

```python
# 3 plots
# create blank canvas
fig = plt.figure()
#set axes
ax1 = fig.add_axes([0,0,1,1]) # full fig
ax2 = fig.add_axes([0.2,0.5,0.25,0.25])
ax3 = fig.add_axes([1.1,0.3,0.25,0.25])

ax1.plot(x,y)
ax1.set_xlabel('X label')
ax1.set_ylabel('Y label')
ax1.set_title('Big Figure')

ax2.plot(x,y)
ax2.set_xlim(8,10)
ax2.set_ylim(4000,10000)
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('Zoomed In')

ax3.plot(a,b)
ax3.set_title('This is third figure')
plt.show()
```

```python
# Figrue parameters
fig = plt.figure(figsize = (10,5), dpi = 100)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)

# Export figure
# fig.savefig('name.png', bbox_inches = 'tight')
```

```python
# Manual spacing on subplots
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12,8))
axes[0][0].plot(a,b)
axes[0][1].plot(b,a)
axes[1][0].plot(x,y)
axes[1][1].plot(y,x)

fig.subplots_adjust(
    left = None,
    bottom = None,
    right = None,
    top = None,
    wspace = 0.2, # width space
    hspace = 0.3  # height space
)

# fig.savefig('name.png', bbox_inches='tight') -> for saving figure
```

## Some advanced plots

### Log scale

- set_yscale()

```python
x = np.arange(6)
fig, axes = plt.subplots(1,2, figsize=(10,4))
axes[0].plot(x, x**2, x, np.exp(x))
axes[0].set_title('Normal Scale')

axes[1].plot(x, x**2, x, np.exp(x))
axes[1].set_yscale('log')
axes[1].set_title('Log scale (y)')
```

### Ticks
- set_xticks
- set_yticks
- set_xticklabels
- set_yticklabels

http://matplotlib.org/api/ticker_api.html 

```python
fig, ax = plt.subplots(figsize=(10,4))
ax.plot(x, x**2, x, x**3, lw=2)

ax.set_xticks([1,2,3,4,5])
ax.set_xticklabels([r'$\alpha$', r'$\beta$', r'$\gamma$', r'$\delta$',
                    r'$\epsilon$'], fontsize = 18)
yticks = [0,50,100,150]
ax.set_yticks(yticks)
ax.set_yticklabels(['$%.1f$' % y for y in yticks], fontsize=18); # latex formatted label
```

```python
# scientific notification
fig, ax = plt.subplots(1,1)
ax.plot(x, x**2, x, np.exp(x))
ax.set_title('Scientific')
ax.set_yticks([0,50,100,150])

from matplotlib import ticker
formatter = ticker.ScalarFormatter(useMathText=True)
formatter.set_scientific(True)
formatter.set_powerlimits((-1,1))
ax.yaxis.set_major_formatter(formatter)
```

### Axis number and label spacing

```python
import matplotlib
# distance btw x and y axis and numbers on axes
matplotlib.rcParams['xtick.major.pad'] = 10
matplotlib.rcParams['ytick.major.pad'] = 10

fig, ax = plt.subplots(1,1)
ax.plot(x, x**2, x, np.exp(x))
ax.set_yticks([0,50,100,150])
ax.set_title('label and axis spacing')

# padding btw axis label and axis nums
ax.xaxis.labelpad = 15
ax.yaxis.labelpad = 15

ax.set_xlabel('x')
ax.set_ylabel('y')
```

### Axis spines

```python
fig, ax = plt.subplots()
ax.plot(x,x)

ax.spines['bottom'].set_color('blue')
ax.spines['top'].set_color('red')

ax.spines['left'].set_linewidth(2)

ax.spines['right'].set_color('none')
ax.yaxis.tick_left()
```

### Twin axes

```python
fig, ax1 = plt.subplots()
ax1.plot(x, x**2, lw=2, color='blue')
ax1.set_ylabel(r'area $(m^2)$', fontsize=18, color='blue')
for label in ax1.get_yticklabels():
  label.set_color('blue')

ax2 = ax1.twinx()
ax2.plot(x, x**3, lw=2, color='red')
ax2.set_ylabel(r'volume $(m^3)$', fontsize=18, color='red')
for label in ax2.get_yticklabels():
  label.set_color('red')
```

### Axes where x and y are zero

```python
fig, ax = plt.subplots()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0)) # set position of x spine to x=0

ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))

xx = np.linspace(-0.75, 1., 100)
ax.plot(xx, xx**3)
```

### Colormap and contour figures
http://www.scipy.org/Cookbook/Matplotlib/Show_colormaps

```python
alpha = 0.7
phi_ext = 2*np.pi*0.5

def flux_qubit_potential(phi_m, phi_p):
  return 2 + alpha - 2 * np.cos(phi_p) * np.cos(phi_m) - alpha * np.cos(phi_ext - 2*phi_p)
```

```python
phi_m = np.linspace(0, 2*np.pi, 100)
phi_p = np.linspace(0, 2*np.pi, 100)
X,Y = np.meshgrid(phi_p, phi_m)
Z = flux_qubit_potential(X,Y).T
```

```python
fig, ax = plt.subplots()
p = ax.pcolor(X/(2*np.pi), Y/(2*np.pi), Z, cmap=matplotlib.cm.RdBu,
              vmin=abs(Z).min(), vmax=abs(Z).max())
cb = fig.colorbar(p, ax=ax)
```

```python
fig, ax = plt.subplots()
im = ax.imshow(Z, cmap=matplotlib.cm.RdBu, vmin=abs(Z).min(), vmax=abs(Z).max(), extent=[0,1,0,1])
im.set_interpolation('bilinear')
cb = fig.colorbar(im, ax=ax)
```

```python
fig, ax = plt.subplots()
cnt = ax.contour(Z, cmap=matplotlib.cm.RdBu, vmin=abs(Z).min(), vmax=abs(Z).max(), extent=[0,1,0,1])
```
